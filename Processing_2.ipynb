```python
import xarray as xr
import shutil

DATA_DIR = 'data/new_region'
```

```python
one_dataset = xr.open_dataset(f'{DATA_DIR}/dataset.nc')
```

# Processing: Write multiple nc w/ different variables

##  Variables and Non-Variables

```python
variables = [var for var in one_dataset.variables.keys()]
non_vars = ['lat', 'lon', 'level', 'time']

variables
```

```python
one_dataset[variables[-1]]
```

## Export Variables

```python
for var in variables:
    if var in non_vars:
        continue
    print(f'Exporting {var}... \n')
    path_dir = f'{DATA_DIR}/{var}'
    path_file = f'{path_dir}/{var}.nc'

    shutil.os.mkdir(path_dir)
    one_dataset[var].to_netcdf(path_file)
```
